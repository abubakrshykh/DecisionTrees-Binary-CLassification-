**DATA PREPROCESSING:**

1-CHECKING AND REMOVING ALL NULL VALUES.

2-REMOVING IRRELEVANT COLUMNS LIKE 'NAME' etc.

3-CHANGING COLUMNS TO CATEGORICAL NUMERIC VALUES.

```python
#CHECKING COLUMS OF OUR DATA
data.columns 
```

```python
#CHECKING NULL VALUES IN OUR DATA
data.isnull().sum()
```

```python

#REMOVING COLUMNS THAT HAS MORE THAN 80% NULL VALUES
data = data.drop(['Cabin'], axis=1)
```

```python
data.isnull().sum()
```

```python
#FILLING OTHER NULL values with median i.e (AGE, EMARKED, FARE)
#FILLING THE NULL VALUES BY MEDIAN VALUES
data.Age = data.Age.fillna(data.Age.median())
data.Fare = data.Fare.fillna(data.Fare.median())

#THIS COLUMN HAS NO NUMERICAL VALUES SO WE CANT PUT MEDIAN IN BLANK SPACES
data.Embarked = data.Embarked.fillna('S')


#CHECKING AGAIN FOR ANY NULL VALUES IN DATA
data.isnull().sum()
```

```python
#CHECKING THE DATSET
data
```

```python
#REMOVING IRRELEVANT COLUMNS/FEATURES
data = data.drop(['PassengerId','Name', 'Ticket'], axis=1) #axis 1 is to actually remove the whole column
```

```python
data
```

```python
#now we have two columNs to change that are (SEX & EMBARKED).
#WE WANT TO CONVERT THEM TO CATEGORICAL VALUES TO FEED INTO OUR MODEL. FOR EXAMPLE (MALE IS 1 AND FEMALE IS 0)

#For changing values to numericals categories we can use LabelEncoder library drom sklearn OR MANUALLY LIKE HERE:

#CHANGING TWO COLUMN VALUES TO CATEGORICAL & NUMERICAL VALUES
data.Embarked = data.Embarked.map( {'S':0,'C':1,'Q':2} )
data.Sex = data.Sex.map( {"male":0, "female":1} )
```
